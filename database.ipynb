```python
import sqlite3
import pandas as pd

con = sqlite3.connect('food_calories.db')
wb = pd.read_excel('food-calories.xlsx',sheet_name = None)

for sheet in wb:
    wb[sheet].to_sql(sheet,con,index=False, if_exists = 'append')
con.commit()
con.close()
```
